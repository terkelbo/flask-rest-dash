```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegressionCV
from sklearn.externals import joblib
```

```python
data = load_iris()
```

```python
feature_names = data['feature_names']
target_names = data['target_names']

X = data['data']
y = data['target']
```

```python
X[0]
```

```python
import json
json.dumps({f:v for f, v in zip(feature_names, X[0])})
```

```python
clf = LogisticRegressionCV(Cs=50,multi_class='multinomial', cv=5, max_iter=500, refit=True)
clf.fit(X, y)
```

```python
joblib.dump(clf, r'model.pkl')
```

```python
joblib.dump(feature_names, 'feature_names.pkl')
joblib.dump(target_names, 'target_names.pkl')
```

```python
clf.predict_proba(X).tolist()
```
